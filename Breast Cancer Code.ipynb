```python
breastcanser_data=data_raw.copy()   #Make a copy of this object's indices and data.
breastcanser_data.head()
```

```python
# Apply Log Transformation to deal with exponential relationship
log_perimeter_mean = np.log(breastcanser_data['perimeter_mean'])
breastcanser_data['perimeter_mean']=log_perimeter_mean #Add new transform data point into the dataframe
breastcanser_data.head()
```

```python
# Apply Log Transformation to deal with exponential relationship
log_area_mean = np.log(breastcanser_data['area_mean'])
breastcanser_data['perimeter_mean']=log_area_mean #Add new transform data point into the dataframe
breastcanser_data.head()
```

```python
breastcanser_data=breastcanser_data.drop(['perimeter_mean'], axis=1) # to remove Log_Price attribute
breastcanser_data.head()
```

```python
breastcanser_data=breastcanser_data.drop(['area_mean'], axis=1) # to remove Log_Price attribute
breastcanser_data.head()
```

```python
#Create dummy variables for categorical variable
data_with_dummy = pd.get_dummies(breastcanser_data)
#data_with_dummy = pd.get_dummies(data_clean,drop_first=True)
data_with_dummy.head(20)
```

```python
data_with_dummy.drop([col for col in data_with_dummy if '_worst' in col], axis=1, inplace=True)
```

```python
data_with_dummy.drop([col for col in data_with_dummy if 'concave' in col], axis=1, inplace=True)
```

```python
data_with_dummy.drop([col for col in data_with_dummy if 'concavity' in col], axis=1, inplace=True)
```
